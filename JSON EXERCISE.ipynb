```python
import json
```

```python
from pandas.io.json import json_normalize
```

```python
#load json as string
data = json.load((open('world_bank_projects.json')))
```

```python
#use normalization to create a table from nested element
data_1 = json_normalize(data, 'mjtheme_namecode','countryname')
data_1
```

```python
#count the number of projects by each country and return 10 countries with the most project
data_1['countryname'].value_counts().head(10)
```

```python
#delete the countryname from the dataframe in order to workwith the project themes
del data_1['countryname']
data_1
```

```python
#filter out all the rows where the name value is '' 
#in order to create a dictionary for the code and the equivalent name
data_2=data_1[data_1.name!='']
#drop all the duplicates
data_2.drop_duplicates()
```

```python
#create a dictionary mapping each project code to its correstponding name
data_dict = dict(zip(data_2.code, data_2.name))
data_dict
```

```python
#create a new column from the data frame mapping all the codes to its corresponding name using the dictionary created
data_1['FillingName'] = data_1['code'].map(data_dict)
```

```python
#using value counts to print the 10 most common project themes
data_1['FillingName'].value_counts().head(10)
```
