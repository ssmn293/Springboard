```python
import requests
```

```python
# Store the API key as a string - according to PEP8, constants are always named in all upper case
API_KEY = 'private'
```

```python
#Set the url to the desired dataset
url = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X.json?private'
```

```python
#Set the parameters for the start date and the end date of the dataset so data's date falls within 2017
params = dict(key = API_KEY, start_date = '2017-01-01', end_date = '2017-12-31')
```

```python
#Request for the desired data of the AFX_X dataset
res =  requests.get(url, params = params)
```

```python
#Convert the data object into a python dictionary
json_data = res.json()
```

```python
#QUESTION 2:
#Create data_keys for the dictionary
data_keys = json_data['dataset']['column_names']
#Get the data from the dataset
data = json_data['dataset']['data']
#transpose the data in order to combine into a dictionary
data_values = [[row[i] for row in data] for i in range(10)]
#Create a dictionary
dataDict=dict(zip(data_keys, data_values))
```

```python
print(dataDict)
```

```python
#QUESTION 3: 
#Find the maximum opening:
#Create a list of Opening price and loop to remove the None data
Opening_price = dataDict['Open']
#Using the for loop to remove all items that are float
for a in Opening_price:
    if type(a) != float:
        Opening_price.remove(a)
```

```python
#Find the maximum Open
Open_max = max(Opening_price)
#Find the minimum Open
Open_min = min(Opening_price)
print('The highest opening price is: ',  Open_max, ' and the lowest is: ',  Open_min)
```

```python
#QUESTION 4:
#Using list comprehension to create a list for a change in price
change = [High - Low for (High, Low) in zip(dataDict['High'], dataDict['Low'])]
#Finding the largest change in any one day
change_max = max(change)
print('The largest change in any one day is: ', change_max)
```

```python
#QUESTION 5:
#Create a list of all the closing price
list1 = dataDict['Close']
#Finding the largest changes in any two days which is the difference the maximum and the minimum
change_max_close = max(list1) - min(list1)
print('The largest change in any two days Closing price is: ', change_max_close)
```

```python
#QUESTION 6:
#Create a list of daily traded volume
list2 = dataDict['Traded Volume']
#Finding the average of daily traded volume by dividing the sum by the length
average = sum(list2)/len(list2)
print('The average daily traded volume is: ', average)
```

```python
#QUESTION 7:
#Sort the daily traded volume in ascending order
Volume_ascend = sorted(list2)
#Finding the length of the list
length=len(Volume_ascend)
print(length)
```

```python
#Finding the location of the median
loc_median = int((length+1)/2)
print(loc_median)
#Finding the median
median_volume = Volume_ascend[loc_median]
print(median_volume)
```
